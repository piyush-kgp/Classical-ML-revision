```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MahalanobisDist(data):
    cov = np.cov(data.T)
    inv_cov_matrix=np.linalg.inv(cov)
    vars_mean=np.mean(data, axis=0)
    diff = data - vars_mean
    md = []
    for i in range(len(diff)):
        md.append(np.sqrt(diff[i].dot(inv_cov_matrix).dot(diff[i])))
    return md
```

```python
X = np.random.normal(50, 100, size=(10000, 2))
MD = MahalanobisDist(data=X)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:,0], X[:, 1], c=MD, s=2, cmap=plt.cm.nipy_spectral)
```

```python
def MahalanobisDistVect(data):
    cov = np.cov(data.T)
    inv_cov_matrix=np.linalg.inv(cov)
    vars_mean=np.mean(data, axis=0)
    diff = data - vars_mean
    md = np.apply_along_axis(lambda d: np.sqrt(d.dot(inv_cov_matrix).dot(d)), 1, diff)
    return md
```

```python
MD = MahalanobisDistVect(data=X)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:,0], X[:, 1], c=MD, s=2, cmap=plt.cm.nipy_spectral)
```
