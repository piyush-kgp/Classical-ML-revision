```python
var_types = {"satisfaction_level": "continuous",
            "last_evaluation": "continuous",
            "number_project": "continuous",
            "average_montly_hours": "continuous",
            "time_spend_company": "continuous",
            "Work_accident": "categorical",
            "promotion_last_5years": "categorical",
            "sales": "categorical",
            "salary": "categorical",
            "left": "categorical"
            }
```

```python
class Node:
    def __init__(self, df, split_var, split_val):
        self.df = df
        # split var and val are needed for prediction
        self.split_var = split_var
        self.split_val = split_val
        self.left = None
        self.right = None
        self.isTerminal = False
        self.y_class = None # fill this for terminal nodes
        
```

```python
class DecisionTree:
    def __init__(self, train_data, target_var, var_types, max_depth=6):
        self.train_data = train_data
        self.target_var = target_var
        self.var_types = var_types
        self.max_depth = max_depth
        
    
    def entropy(self, pd_series):
        entropy_val = 0
        N = len(pd_series)
        for value in pd_series.unique():
            count = sum(pd_series==value)
            entropy_val+= -count/N*np.log2(count/N)
        return entropy_val


    def information_gain(self, df, feature, feature_val, feature_val_type, target):
        count_all = df.shape[0]
        initial_entropy = self.entropy(df[target])

        if feature_val_type=="categorical":
            subset_df1 = df[df[feature]==feature_val]
            subset_df2 = df[df[feature]!=feature_val]
        elif feature_val_type=="continuous":
            subset_df1 = df[df[feature]<=feature_val]
            subset_df2 = df[df[feature]>feature_val]
            
        count1, count2 = subset_df1.shape[0], subset_df2.shape[0]

        subset_df1_entropy = self.entropy(subset_df1[target])
        subset_df2_entropy = self.entropy(subset_df2[target])

        final_entropy = count1/count_all*subset_df1_entropy + count2/count_all*subset_df2_entropy

        return initial_entropy - final_entropy    


    def get_split_var_and_split_val(self, df):
        feature_columns = list(df.columns)
        feature_columns.remove(self.target_var)
        inf_gains = {}
        for feature in feature_columns:
            if self.var_types[feature]=="categorical":
                split_values = df[feature].unique()
            elif self.var_types[feature]=="continuous":
                unique_values = sorted(df[feature].unique())
                split_values = [0.5*(unique_values[i-1]+unique_values[i]) \
                                for i in range(1, len(unique_values))]
            
            for value in split_values:
                inf_gain = self.information_gain(df, feature, value, self.var_types[feature], self.target_var)
                inf_gains[(feature, value)] = inf_gain
        
        split_var, split_val = max(inf_gains, key=lambda a: inf_gains[a])
        return split_var, split_val             
 

    def createTree(self, root, depth):
        print(root.df.shape, depth)
        if depth>=self.max_depth or root.df.shape[0]==1:
            root.isTerminal = True
            root.y_class = sum(root.df[self.target_var]==1)/root.df.shape[0] > 0.5
            return

        split_var, split_val = self.get_split_var_and_split_val(root.df)
        print(split_var, split_val)
        if self.var_types[split_var] == "categorical":
            df_left = df[df[split_var]==split_val]
            df_right = df[df[split_var]!=split_val]
        elif self.var_types[split_var] == "continuous":
            df_left = df[df[split_var]<=split_val]
            df_right = df[df[split_var]>split_val]
        root.left = Node(df_left, split_var, split_val)
        root.right = Node(df_right, split_var, split_val)
        self.createTree(root.left, depth+1)
        self.createTree(root.right, depth+1)

        
    def fit(self):
        split_var, split_val = self.get_split_var_and_split_val(self.train_data)
        self.root = Node(self.train_data, split_var, split_val)
        self.createTree(self.root, 1)
        
        
    def predictUtil(self, root, sample):
        if root.isTerminal:
            return int(root.y_class)
        if self.var_types[root.split_var]=="categorical":
            isLeft = sample[root.split_var]==root.split_val
        elif self.var_types[root.split_var]=="continuous":
            isLeft = sample[root.split_var]<=root.split_val
        if isLeft:
            return self.predictUtil(root.left, sample)
        else:
            return self.predictUtil(root.right, sample)
            
    def predict(self, batch_df):
        preds = []
        for i in batch_df.index:
            sample = batch_df.loc[i, :]
            pred = self.predictUtil(self.root, sample)
            preds.append(pred)
        return preds
```

```python
decision_tree = DecisionTree(train_data = train, target_var="left", var_types=var_types)
decision_tree.fit()
```
